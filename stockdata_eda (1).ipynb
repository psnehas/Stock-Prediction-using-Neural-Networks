```python
list1 = ['AAPL','AMZN','ALXN','AMGN']
plt.figure(figsize=(14, 12))
# plt.subplots_adjust(top=1.25, bottom=)
for i,v in enumerate(list1,1):
    df_plot =  df[df['Ticker'] == v]
    plt.subplot(2, 2, i)
    x = df_plot['Date']
    y = df_plot['Adj Close']
    plt.plot(x,y)
    plt.xlabel('Date')
    plt.ylabel("Adj Close")
    plt.title(v)
    plt.show
```
