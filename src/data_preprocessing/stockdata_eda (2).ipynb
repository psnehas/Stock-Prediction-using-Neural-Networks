```python
# Stock Market Data Analysis. The stock data is for the period between 2000 and 2020
```

```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
%matplotlib inline
```

```python
df = pd.read_csv("stockdata.csv")
```

```python
df.shape
```

```python
df.head()
```

```python

from datetime import datetime
df['Date']=pd.to_datetime(df['Date'])
```

```python
df.dtypes
```

```python
#Filter the dataframe to analyse the stock data for a certain period
filtered_df=df.query("Date >= '2020-01-01' and Date <='2020-12-31'")
filtered_df.head()
```

```python
filtered_df.shape
```

```python
list1 = ['AAPL','AMZN','ALXN','AMGN']
plt.figure(figsize=(20, 12))
# plt.subplots_adjust(top=1.25, bottom=)
for i,v in enumerate(list1,1):
    df_plot =  filtered_df[filtered_df['Ticker'] == v]
    plt.subplot(2, 2, i)
    x = df_plot['Date']
    y = df_plot['Close']
    plt.plot(x,y)
    plt.xlabel('Date')
    plt.ylabel("Close")
    plt.title(v)
    plt.show

```
